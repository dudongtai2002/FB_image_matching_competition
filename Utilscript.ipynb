```python
import numpy as np
def create_embedding_dic(n):
    sampl = np.random.uniform(low=0.5, high=13.3, size=(512,))
    result = {}
    for i in range(n):
        sampl = np.random.uniform(low=0., high=2.3, size=(512,))
        result['R'+ str(i)] = sampl
    return result

small = create_embedding_dic(500)
print(len(small))
large = create_embedding_dic(1000)
```

```python
def convert_to_dict(dict1,dict2):
    #dict1 = pd1.set_index('image_id').to_dict()['embedding']
    #dict2 = pd2.set_index('image_id').to_dict()['embedding']
    #dict1 = pd1.to_dict()['embedding']
    #dict2 = pd2.to_dict()['embedding']
    result = {}
    i=0
    for image_id1, embedding1 in dict1.items():
        neighbor_image_id = ""
        min_distance = float('inf')
        for image_id2, embedding2 in dict2.items():
            distance = np.linalg.norm(embedding1-embedding2)
            if distance < min_distance:
                neighbor_image_id = image_id2
                min_distance = distance
        result[image_id1] = (neighbor_image_id, min_distance)
        i+=1
        if(i%50==0):
            print("finish:{}".format(i))
    return result
```

```python
from timeit import default_timer as timer
from datetime import timedelta
start = timer()
result1 = convert_to_dict(small, large)
end = timer()
print("finish method1, takes: {}".format(timedelta(seconds=end-start)))
print(len(result1))
```

```python
def show_pairwise_distance(query_id, ref_id, df_query, df_ref):
    query_emd = df_query.loc[query_id]['embedding']
    ref_emd = df_ref.loc[ref_id]['embedding']
    return np.linalg.norm(query_emd-ref_emd)
```
